```python
# Load data
with open("../data/day9.txt", "r") as f:
    input = f.read()

data = [int(value) for value in input.split("\n")]
data
```

```python
# Puzzle 1
from itertools import combinations 
    
line = 25
for n in range(25, len(data)):
    print(data[n])
    previous_25 = data[n-25:n]

    pairs = combinations(previous_25, 2) 

    finished = False
    for pair in pairs:
        finished = sum(pair) == data[n]
        if finished:
            break
    print(finished, pair)
    
    if not finished:
        print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        
```
