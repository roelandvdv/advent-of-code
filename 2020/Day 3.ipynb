```python
import re

input_n = []
for line in input:
    input_n.append([m.start() for m in re.finditer('#', line)])
    
input_n
```

```python
pos_x = 0
tree_count = 0

for line in input_n:
    if (pos_x % 31) in line:
        tree_count += 1
    pos_x += 3

print(tree_count)
```

```python
# Generalize to skip m lines

steps_x = [1, 3, 5, 7, 1]
steps_y = [1, 1, 1, 1, 2]

trees = []
for (x, y) in zip(steps_x, steps_y):
    pos_x = 0
    pos_y = 0
    tree_count = 0

    for line in input_n:
        if pos_y % y == 0:
            if (pos_x % 31) in line:
                tree_count += 1
            pos_x += x

        pos_y += 1

    trees.append(tree_count)

print(trees)
```
